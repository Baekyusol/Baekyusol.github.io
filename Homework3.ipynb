### 리스트에는 변수를 그대로 대입할 수 없다. 키워드나 함수를 사용하거나 리스트를 대입해야 한다. 

```python
L=[1,2,3,4,5]
L[1:3]=(100)
print(L)
```

### 하나의 항목으로 된 튜플을 만들려면 값 뒤에 쉼표를 붙여서 완성해야 한다. 위의 괄호는 수학적 괄호, 즉 우선순위를 위한 괄호로서 작용하므로 리스트에 대입할 수 없다. 

```python
L=[1,2,3,4,5]
L[1:3]=(100,101,102)
print(L)
```

```python
L=[1,2,3,4,5]
L[1:3]=[100,101,102]
print(L)
```

```python
L=[1,2,3,4,5]
L[1:3]=[[100,101,102]]
print(L)
```

```python
s = '  first star   :   second star   :    third star  '
s1=s.split(":")
L=[]
for i in s1:
    i=i.strip()
    L.append(i)
print(L)
```

### for ~in문을 이용하여 작성한 코드이다. 일단 문자열 s를 : 를 기준으로 나누어 리스트에 저장한다. 그리고 새로운 리스트를 만들고 여기에 strip 메소드를 활용해 앞뒤의 공백을 지워준것을 저장한다. 그리고 출력한다. 

```python
s = '  first star   :   second star   :    third star  '
s1=s.split(":")
L=[l.strip() for l in s1]
print(L)
```

### 위와 :를 기준으로 나누어주는 것은 똑같다. 여기에서 새로운 리스트를 만들 때 바로 리스트 내포를 이용하여 리스트를 반복해주고 strip 메소드를 이용해 앞뒤 공백을 지워준 것을 저장한다. 그리고 출력한다. 

```python
def addall(list):
    sum=0
    for i in list:
        if i>0:
            sum+=i
    print(sum)
    
addall([1])
```

### 먼저 for ~in문을 이용하면 일단 매개변수로 받은 리스트를 범위로 설정하고 if문으로 0보다 큰 정수라고 범위를 지정해준다음, sum변수에 차근차근 더해준다. 

```python
def addall(list):
    L=[i for i in list if i>0]
    print(sum(L))
    
addall([1,2,3,4,5,6,7,8,9])
```

### 이번엔 리스트 내포를 이용하여 리스트를 선언하고 바로 i가 0보다 큰 정수라면 i를 저장해준다. 그리고 출력할 때 내장함수를 이용하여 합을 출력해준다.

```python
def addallodd(list):
    sum=0
    for i in list:
        if i%2==1:
            sum+=i
    print(sum)
    
addallodd([1])
```

### 위와 정확히 똑같지만 이 경우에는 if문에서 홀수로 범위를 지정해주어야 한다. 따라서 리스트를 반복했을 때 i가 2로 나눈 값이 1이면 홀수이므로 식을 써준 다음 출력해준다. 

```python
def addallodd(list):
    L=[i for i in list if i%2==1]
    print(sum(L))
    
addall([1,2,3,4,5,6,7,8,9])
```

### 마찬가지로 리스트 내포를 이용해 작성한 줄에서 if문을 홀수로 나오도록 써준다음 출력해준다.

```python
def length(X):
    l=len(str(X))
    return l

def isStrange(num):
    Strange=False
    tries=1
    if length(num)==1:
        Strange=True
    elif length(num)>1:
        if num%length(num)==0:
            if num/length(num)==Strange:
                Strange=True
        else:
            return False
    else:
        return False
    
    print('{}. sum = {}  => {}'.format(tries,num,Strange))
    tries+=1
    isStrange((int)(input()))
    return True
    

isStrange(0)
```

### 일단 length함수에서는 len은 int를 받을 수 없다고 오류가 나서 str로 바꿔주고 길이를 반환해주었다. isStrange함수는 일단 Strange를 기본 False로 선언해준다. 그리고 length함수를 실행했을때 그 값이 1일 때, 그리고 1 이상이었을 경우 나누어 떨어지고 나누었을때 값이 False 즉 Strange 값이 나오면 Strange값을 True로 바꿔주었다. 그러나 한자리수는 되지만 두자리수는 되지 않았다. 재귀함수를 부를 때 문자가 있는건가 생각해 여러가지로 시도해보았지만 오류만 뜰뿐이었다. 그리고 tries도 더해지는 것으로 하고 싶었지만 계속 1로 초기화되었다...

```python
import requests
import re
import string

req=requests.get('https://www.naver.com/')

source=req.text

#print(source)
body=re.search('<body.*/body>',source,re.I|re.S)

if(body is None):
    print("No <body> in html")
    exit()
    
body=body.group()
body=re.sub('<script.*?.*?</script>','',body,0,re.I|re.S)
text=re.sub('<.+?>','',body,0,re.I|re.S)
others=re.sub('{.+?}','',text,0,re.I|re.S)
others2=re.sub('._1syGnXOL.*?.*?_:after','',others,0,re.I|re.S)
nospace = re.sub('-|@|\t|\r|\n', '', others2)
nopunc=nospace.translate(str.maketrans('', '', string.punctuation))

result=nopunc.split(' ')
result = list(map(lambda x: x.strip(), result))
result = list(filter(lambda x: x != '', result))
count=dict()
for word in result:
    count[word]=count.get(word,0)+1
print(count)
```

### 리스트에서 양쪽 공백과 빈 항목을 제거해준 것까지는 그 전 과제와 동일하다. 구두문자를 제거해주기 위해 translate를 사용하여 모든 구두문자들을 빈칸으로 바꾸어 주었다. 이는 리스트에서는 못 쓰니 리스트로 만들기 전에 작성해주었다. 그리고 count라는 딕셔너리를 만들어준 다음 for ~in문을 사용해 리스트를 반복해준다음 count 안에 단어를 넣어준다. 이때 get메소드를 이용하여 count내에 word가 없다면 0을 반환해주고 1을 더해주어 출현빈도를 나타내주었다.
